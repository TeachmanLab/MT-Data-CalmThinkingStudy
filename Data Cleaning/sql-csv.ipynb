```python
from datetime import datetime
import mysql.connector
import pandas as pd
import os


dl_date = datetime.today().strftime('%d_%m_%Y')


saveDir = "/Data/CalmThinking-"+ dl_date
if not os.path.exists(saveDir):
    os.makedirs(saveDir)

# change the current directory
os.chdir(saveDir)

host = "hostConnection"
user = "userName"
password= "passWord"
database = "calm"
auth_plugin= "mysql_native_password"

mydb = mysql.connector.connect(
    host=host,
    user=user,
    password= password,
    database=database,
    auth_plugin = auth_plugin
)

mycursor = mydb.cursor()

## extract all the tables in this dataset
query = "show tables;"
tables = pd.read_sql_query(query,mydb)

#use a special character for spliting the table name from the date
split_char = "-"

## save the tables as a csv format
for tbl in tables.values:
    tblName = tbl
    query = "select * from {};".format(tblName[0])
    data = pd.read_sql_query(query,mydb)
    print("--------------------------------------")
    print("The name of the table is: {} \n with {} number of columns which are \n{}".format(tblName, len(data.columns), data.columns.values))
    print("\nSaving {} to a csv format.".format(tblName[0]))
    data.to_csv(tblName[0]+ split_char + dl_date + ".csv")
    print("Done!\n")

print("\n--------------------------------------")
print("--------------------------------------")
print("There should be {} number of tables in {} folder".format(len(tables.values), "CalmThinking-"+ dl_date))
print("--------------------------------------")
print("--------------------------------------")
```
