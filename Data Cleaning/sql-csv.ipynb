```python
from datetime import datetime
import mysql.connector
import pandas as pd
import os


os.chdir("/Data)

dl_date = datetime.today().strftime('%d-%m-%Y')

host = "hostConnection"
user = "userName"
password= "passWord"
database = "calm"
auth_plugin= "mysql_native_password"

mydb = mysql.connector.connect(
    host=host,
    user=user,
    password= password,
    database=database,
    auth_plugin = auth_plugin
)

mycursor = mydb.cursor()

## extract all the tables in this dataset
query = "show tables;"
tables = pd.read_sql_query(query,mydb)

## save the tables as a csv format
for tbl in tables.values:
    tblName = tbl
    print(tblName)
    query = "select * from {};".format(tblName[0])
    data = pd.read_sql_query(query,mydb)
    data.to_csv(tblName[0]+ "_" + dl_date + ".csv")
    
```
